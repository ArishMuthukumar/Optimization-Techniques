**7. Conjugate Directions (with provided directions)**

```python
import numpy as np

# Quadratic data
Q = np.array([[4, 2],
              [2, 2]], dtype=float)
b = np.array([-1, 1], dtype=float)

# Objective
def f(x):
    return 0.5 * (x @ Q @ x - x @ b)

# Gradient
def grad(x):
    return Q @ x - 0.5 * b

# Initial point
x0 = np.array([0.0, 0.0])

# Directions
d0 = np.array([1.0, 0.0])
d1 = np.array([-3/8, 3/4])

# Conjugate direction iterations
x = x0.copy()
print("Initial x:", x, "f(x):", f(x))

for k, d in enumerate([d0, d1]):
    g = grad(x)
    alpha = (g @ d) / (d @ Q @ d)
    x = x - alpha * d
    print(f"Step {k}: alpha={alpha:.6f}, new x={x}, f(x)={f(x):.6f}")

# Compare to exact solution (solve Qx = 0.5 b)
x_exact = np.linalg.solve(Q, 0.5 * b)
print("Exact minimizer:", x_exact)
```
